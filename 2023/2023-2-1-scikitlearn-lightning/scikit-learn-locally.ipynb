```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
random_state = 1234
```

```python
# Load the iris dataset
iris = load_iris()
X, Y = iris.data, iris.target

print(X.shape)
```

```python
feature_names = iris["feature_names"]
target_names = iris["target_names"]

print(feature_names)
print(target_names)
```

```python
for i, target_name in enumerate(target_names):
    x = iris.data[Y == i][:, 2:]
    y = Y[Y == i]

    plt.scatter(x[:, 0], x[:, 1], label=target_name)

plt.legend(target_names)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("Iris Flower data distribution")
plt.show()
```

```python
# Split the dataset into training and test set
X_train, X_testt, y_train, y_test = train_test_split(X, Y, random_state=random_state)

# initialize a Decision tree model
clf = tree.DecisionTreeClassifier(random_state=random_state)

# train the model
clf = clf.fit(X_train, y_train)

# check accuracy
print(f"train accuracy: {clf.score(X_train, y_train)}")
print(f"test accuracy: {clf.score(X_test, y_test)}")
```

```python
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
